## Introduction


This project analyzes PBMC single-cell RNA sequencing data using Scanpy.

PBMC are cells known as Peripheral Blood Mononuclear Cells. These cells include lymphocytes and monocytes. They are important in fields of immunology, vaccine development, and cancer immunotherapy research. They share a characteristic of being a diverse set of immune cells in human blood. They have a single round nucleus.

The goal is to identify groups of similar immune cells using preprocessing, PCA, UMAP, and clustering.

## Imports

Scanpy is a python tool that aids in analyzing single-cell gene expression data.
MatPlotLib is a python library used for plotting visualizations.

```python
import scanpy as sc
import matplotlib.pyplot as plt
```

## Data

For this project, I will be using the 3k PBMCs dataset from 10x Genomics. This is a useful dataset used to learn single-cell RNA sequencing. It contains about 2,700 PBMCs sequenced on an Illumina NextSeq 500 platform, which is a machine that uses SBS chemistry to read genetic codes.

```python
data = sc.datasets.pbmc3k()
data
data.obs.head()
data.var.head()
data.X[:5, :5].toarray()
```

## Filter Data

Next, we must filter the data. This is because we do not want to keep cells that do not have very much genes, or genes that are not present in many cells. A good starting point is to keep only cells that have more than 200 genes expressed, and keep only genes that are expressed in at least 3 cells.


```python
sc.pp.filter_cells(data, min_genes=200)
sc.pp.filter_genes(data, min_cells=3)
data
```

## Normalize Data

After filtering, we normalize the data because cells can have different total RNA counts even if they are the same cell type. This variation can come from biological factors such as cell state, environment, or cell cycle stage, as well as technical factors like differences in sequencing depth or RNA capture efficiency.

Normalization adjusts the expression values so that each cell has the same total count, allowing gene expression to be compared based on relative proportions rather than raw totals. This prevents cells with higher RNA counts from appearing to have artificially higher gene expression.


```python
#every cell has the same total RNA
#gene values represent proportions, not raw amounts
sc.pp.normalize_total(data)

#Even after normalization, gene values can have a huge range. compresses large values
#keeps small differences visible
#reduces the impact of extreme genes
sc.pp.log1p(data)
```

We keep highly variable genes because they are the genes that best distinguish different cell types from each other. Genes that are expressed similarly in every cell do not help distinguish cell types, while highly variable genes can reveal meaningful biological differences that PCA and clustering can use. Rather than similarities, we are looking for differences.
